```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time

#동적 페이지 접근
driver =webdriver.Chrome()
```

```python
driver.get('https://search.shopping.naver.com/search/all?query=%EB%8B%AD%EA%B0%80%EC%8A%B4%EC%82%B4')
```

```python
#페이지 소스 불러오기

#무한 스크롤 처리
previous_height = driver.execute_script("return document.body.scrollHeight")

    
for i in range(1,5):
    #스크롤 끝까지 내리기
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(1)
    now_height = driver.execute_script("return document.body.scrollHeight")
    if previous_height == now_height:
        break
    previous_height = now_height
html = driver.page_source    
soup = BeautifulSoup(html, 'html.parser')

products = soup.select('.product_item__MDtDF')
```

```python
for product in products:
    name = product.select_one('.product_title__Mmw2K').text
    link = product.select_one('.product_title__Mmw2K >a').attrs['href']
    price = int(product.select_one('.price_num__S2p_v').text.replace("원" , "").replace(',', ""))
    if product.select_one('[data-nclick*=purchase] em'):
        purchase = product.select_one('[data-nclick*=purchase] em').text.replace(',' ,'')
        if '만' in purchase:
            purchase = int(float(purchase.split('만')[0])*10000)
    else:
        purchase = ''
    print(name, link , price , purchase)
```
